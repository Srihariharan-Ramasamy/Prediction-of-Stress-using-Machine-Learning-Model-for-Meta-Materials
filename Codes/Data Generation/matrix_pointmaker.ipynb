```python
import os
import numpy as np
import matplotlib.pyplot as plt

def cal_dist(p1,p2):
        dist = ((p2[0]-p1[0])**2+(p2[1]-p1[1])**2)**0.5
        return dist

N=10
OUT_DIR = "MLMM_Coords_run3"
os.makedirs(OUT_DIR, exist_ok=True)
# If values are comma-separated
for fileNum in range(1,10001):
    inputfName = f"MLMM_stress_run3/cell_{fileNum:05d}.csv"
    matrix = np.loadtxt(inputfName, delimiter=",", dtype=int)

    #print(A.shape)
    #print(A)
    x=[]
    y=[]

    for i in range(0,N):
        for j in range(0,N-1):
            
            if matrix[i,j]+matrix[i,j+1]==1:
                y.append(-(i)*0.1)
                y.append(-(i)*0.1-0.05)
                y.append(-(i)*0.1-0.1)
                x.append((j+1)*0.1)
                x.append((j+1)*0.1)
                x.append((j+1)*0.1)
        if matrix[i,0]==1:
            y.append(-i*0.1)
            y.append(-(i*0.1)-0.05)
            y.append(-(i*0.1)-0.1)
            x.append(0)
            x.append(0)
            x.append(0)
        if matrix[i,9]==1:
            y.append(-i*0.1)
            y.append(-(i*0.1)-0.05)
            y.append(-(i*0.1)-0.1)
            x.append(1)
            x.append(1)
            x.append(1)

    for j in range(0,N):
        for i in range(0,N-1):
            
            if matrix[i,j]+matrix[i+1,j]==1:
                y.append(-(i+1)*0.1)
                y.append(-(i+1)*0.1)
                y.append(-(i+1)*0.1)
                x.append((j)*0.1)
                x.append((j)*0.1+0.05)
                x.append((j)*0.1+0.1)
        if matrix[0,j]==1:
            y.append(0)
            y.append(0)
            y.append(0)
            x.append((j*0.1))
            x.append((j*0.1)+0.05)
            x.append((j*0.1)+0.1)
        if matrix[9,j]==1:
            y.append(-1)
            y.append(-1)
            y.append(-1)
            x.append((j*0.1))
            x.append((j*0.1)+0.05)
            x.append((j*0.1)+0.1)
        
    if fileNum==10000:
        print(matrix)
        plt.scatter(x, y, c='blue', marker='o')  # c sets color, marker='o' makes circles
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.xlim(-0.2, 1.2)
        plt.ylim(-1.2, 0.2)

        plt.xticks([i/10 for i in range(10)])  # 0.0, 0.1, ..., 0.7
        plt.yticks([-i/10 for i in range(10)])
        plt.title("Scatter plot of points")
        plt.grid(True)
        #plt.axis("equal")   # keeps aspect ratio square
        plt.show()

    # x and y are lists of equal length
    pairs = [list(p) for p in zip(x[1:], y[1:])]   # [[x[1], y[1]], [x[2], y[2]], ...]
    # pairs

    a = pairs

    sorted_a = [a[0]]
    a = a[1:]
    while len(a):
        pts_dist = []
        p1 = sorted_a[-1]
        for v in a:
            pts_dist.append(cal_dist(p1,v))
            q = pts_dist.index(min(pts_dist))
        sorted_a.append(a[q])
        a.remove(a[q])
    sorted_a.append(sorted_a[0])
    # print(sorted_a)
    
    outputfName = f"coords_{fileNum:05d}.csv"
    print(fileNum, outputfName)
    
    np.savetxt(os.path.join(OUT_DIR, outputfName), np.asarray(sorted_a, float),
            delimiter=",", fmt="%.6f")
```
